# Questão 1:

Adaptação dos eixos para encaixar no formato 200, 200

```python
def curve1param(t):
    return (round(100*math.cos(t))+100, 200 - (round(100*math.sin(t))+100))
```

Função auxiliar que verifica se determinado ponto encontra-se dentro dos limites da imagem

```python
def insideImg(point):
  x = point[0]
  y = point[1]
  return x > 0 and x < 200 and y > 0 and y < 200
```

```python
img1 = Image.new('1', (200,200), 0)

for i in range(360):
    pixel = curve1param (i)
    x = pixel[0]
    y = pixel[1]
    if insideImg(pixel):
        img1.putpixel(pixel, 1)

plt.imshow(img1)
```

# Questão 2

Adaptação dos eixos para encaixar no formato 200, 200

```python
def curve2param(t):
    return (round(t*math.sin(t)) + 100, 200 - (round(t*math.cos(t)) + 100))
```

## A)

```python
img2a = Image.new('1', (200,200), 0)
t=0
while t<100:
    pixel = curve2param(t)
    if insideImg(pixel):
        img2a.putpixel(pixel, 1)
    t+=0.1

plt.imshow(img2a)
```

## B)

```python
def incrementSample(t):
  return 1/math.sqrt(1+t*t)
```

```python
img2b = Image.new('1', (200,200), 0)
t=0
while t<100:
    pixel = curve2param(t)
    if insideImg(pixel):
        img2b.putpixel(pixel, 1)
    t += incrementSample(t)

plt.imshow(img2b)
```

# Questão 3




Adaptação dos eixos para encaixar no formato 200, 200

```python
def curve3a(x,y):
  y = 200-y
  return x*x + (y-200)*(y-200) <= 200*200

def curve3b(x,y):
  y = 200-y
  return (x-200)*(x-200) + y*y <= 200*200

def curve3c(x,y):
  y = 200 -y
  return x + y > 200
```

```python
img3 = Image.new('1', (200,200), 0)

for x in range(200):
    for y in range(200):
        if curve3a(x,y) and curve3b(x,y) and curve3c(x,y):
            img3.putpixel((x,y), 1)

plt.imshow(img3)
```

# Questão 4

Adaptação dos eixos para encaixar no formato 200, 200

```python
def curve4(x,y):
  x = x/50 - 2
  y = (200-y)/50 - 2
  return y*y - x*x*x + x
```

```python
def isPositive(x):
  return x>0

def sameSignalVertices(v1, v2, v3, v4):
  return isPositive(v1) != isPositive(v2) or isPositive(v1) != isPositive(v3) or isPositive(v1) != isPositive(v4)
```

Analizando os vértices dos pixels da imagem

```python
img4 = Image.new('1', (200,200), 0)

for i in range(200):
  for j in range(200):
    v1 = curve4(i - 0.5, j - 0.5)
    v2 = curve4(i - 0.5, j + 0.5)
    v3 = curve4(i + 0.5, j - 0.5)
    v4 = curve4(i + 0.5, j + 0.5)

    if sameSignalVertices(v1, v2, v3, v4):
      img4.putpixel((i,j), 1)

plt.imshow(img4)
```
